```python
import pandas as pd
PriceAll_Csv = pd.read_csv("C:/Users/Irsyad/Documents/Introduction to Portfolio Construction and Analysis with Python/Python Personal Exercise.csv", sep=";",
                       header=0, index_col=0, parse_dates=True, na_values=-99.99)
```

```python
print(type(PriceAll_Csv))
```

```python
print(PriceAll_Csv.columns)
```

```python
columns = ["BMRI", "BBCA"]  # Adjust to match the DataFrame column names
PriceAll_Csv = PriceAll_Csv[columns]
PriceAll_Csv
```

```python
PriceAll_Csv = PriceAll_Csv.reset_index()  # Resets the index, ADRO becomes a regular column
PriceAll_Csv = PriceAll_Csv[columns]  # Select only BMRI and BBCA
print(PriceAll_Csv)
```

```python
PriceAll_Csv.plot.bar()
```

```python
Returns_BMRIBCA = PriceAll_Csv.pct_change().iloc[1:22]
Returns_BMRIBCA
```

```python
Returns_BMRIBCA.shape
```

```python
Returns_BMRIBCA.std()
```

```python
annualized_vol = Returns_BMRIBCA.std()*np.sqrt(21)
annualized_vol
```

```python
n_period = Returns_BMRIBCA.shape[0]
return_per_semester = (Returns_BMRIBCA+1).prod()**(1/n_period) - 1
return_per_semester
```

```python
annualized_return = (return_per_semester+1)**2-1

#2 because 1 year consist of 2 semester.

annualized_return
```

```python
annualized_return/annualized_vol
```

```python
riskfree_rate = 0.03

excess_return = annualized_return - riskfree_rate
sharpe_ratio = excess_return/annualized_vol

sharpe_ratio
```

It means, adjusted risk to return for BBCA is far more superior than BMRI within 2011-2023 period
